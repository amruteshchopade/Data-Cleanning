# DATA CLEANNING 1

Importing Pandas

```python
# importing libraries
import pandas as pd
import numpy as np
```

Creating Data Frame


```python
# read data
df = pd.read_csv('/content/drive/MyDrive/Data-cleaning-for-beginners-using-pandas.csv')
```

# Creating A copy of data Frame to work

```python
# creatinf copy of main dataframe
df_copy = df.copy()
df_copy
```

```python
#converting all column name to lower and replacinng " " with "_"
df_copy.columns.str.lower().str.replace(" ","_")
```

# Filling missing values in Age Column

```python
#calculating median of column age
Avg_age = df_copy['Age'].median()
print("Median of Colum Age: ",Avg_age)
#filling median of age coloumn in missing value
df_copy['Age'] = df_copy['Age'].fillna(Avg_age)
df_copy['Age'] = df_copy['Age'].astype(int)
df_copy['Age']
```

# **Salary**

```python
# Replace '$' with a space and 'k' with '000'
df_copy['Salary'] = df_copy['Salary'].str.replace('$', '').str.replace('k', '000')

# Strip leading and trailing whitespaces
df_copy['Salary'] = df_copy['Salary'].str.strip()

# Create 'Starting_Salary' and 'Final_Salary' columns
df_copy[['Starting_Salary', 'Final_Salary']] = df_copy['Salary'].str.split('-', 1, expand=True)
df_copy = df_copy.drop(columns=['Salary'])
df_copy['Starting_Salary'] = df_copy['Starting_Salary'].astype(int)
df_copy['Final_Salary'] = df_copy['Final_Salary'].astype(int)
print(df_copy[['Starting_Salary', 'Final_Salary']])
```

# Filling missing value in Rating column

---



```python
Avg_rating = df_copy['Rating'].median()
print("Median of column Rating : ",Avg_rating)
df_copy['Rating'] = df_copy['Rating'].fillna(Avg_rating)
df_copy['Rating'] = df_copy['Rating'].replace(-1, Avg_rating)
df_copy['Rating']
```

# LOCATION

---



```python
#rePlacing " " to ","
df_copy['Location'] = df_copy['Location'].str.replace(' ', ',')

df_copy['Location'] = df_copy['Location'].str.replace('New,York,Ny', 'New York,Ny')

# Split the 'Location' column into two columns 5n based on the comma
df_copy[['Country_Name', 'Initials']] = df_copy['Location'].str.split(',', 1, expand=True)
# If you want to remove leading/trailing whitespaces from 'Initials'
df_copy['Initials'] = df_copy['Initials'].str.strip()
df_copy = df_copy.drop(columns=['Location'])
df_copy['Country_Name'] = df_copy['Country_Name'].astype(str)
df_copy['Initials'] = df_copy['Initials'].astype(str)
# Print the modified DataFrame
print(df_copy[['Country_Name', 'Initials']])
```

# ESTABLISHED YEAR

```python
# Replace -1 with 'Not Available' in the 'Established' column
df_copy['Established_Year'] = df_copy['Established'].replace(-1, 'Not Available')

# Drop the original 'Established' column if needed
df_copy = df_copy.drop(columns=['Established'])

# Display the 'Established_Year' column
print(df_copy['Established_Year'])
```

# easy apply

#Outliers:
**Question**:Identify potential outliers in numerical indicators (e.g., Age, Salary, Rating). Should outliers be removed or adjusted?

```python
numerical_columns = df_copy.select_dtypes(include=np.number).columns
print("Numerical Columns:\n", numerical_columns)
for column in numerical_columns:
  q1 = df_copy[column].quantile(0.25)
  q3 = df_copy[column].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  outliers = df_copy[(df_copy[column] < lower_bound) | (df_copy[column] > upper_bound)]
  print(f"Outliers in {column}:\n", outliers)
  print("-----------------------------------------------------------------------")
```
